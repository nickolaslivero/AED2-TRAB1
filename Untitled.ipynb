```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('metricas.csv')

y_limits = {
    100: (0.000001, 0.0001),
    1000: (0.000001, 0.001),
    10000: (0.00001, 0.01),
    100000: (0.001, 70)
}

array_sizes = [100, 1000, 10000, 100000]
input_types = ['Random', 'Sorted']
algorithms = data['Algorithm'].unique()

for size in array_sizes:
    fig, ax = plt.subplots(figsize=(12, 7))
    ind = np.arange(len(algorithms))
    width = 0.35

    for i, input_type in enumerate(input_types):
        filtered_data = data[(data['Array Size'] == size) & (data['Type'] == input_type)]
        times = filtered_data[filtered_data['Array Size'] == size]['Time taken']

        ax.bar(ind + i * width, times, width, label=f'{input_type}')

    ax.set_xlabel('Algorithms')
    ax.set_ylabel('Time taken')
    ax.set_title(f'Time taken by Algorithms for Array Size {size}')
    ax.set_xticks(ind + width)
    ax.set_xticklabels(algorithms)
    ax.legend()
    plt.xticks(rotation=45)
    
    y_lower_limit, y_upper_limit = y_limits[size]
    ax.set_ylim(y_lower_limit, y_upper_limit)
    
    plt.tight_layout()
    plt.show()
```

```python
algorithms = data['Algorithm'].unique()


plt.figure(figsize=(12, 8))


for algorithm in algorithms:
    algorithm_data = data[data['Algorithm'] == algorithm]
    plt.plot(algorithm_data['Array Size'], algorithm_data['Time taken'], marker='o', label=algorithm, linewidth=2)

plt.title('Comparao do tempo de execuo por tamanho de array para diferentes algoritmos de ordenao')
plt.xlabel('Tamanho do Array')
plt.ylabel('Tempo de execuo (segundos)')
plt.legend()
plt.grid(True, axis='y') 
plt.yscale('log') 

plt.gca().spines['top'].set_visible(False) 
plt.gca().spines['right'].set_visible(False) 
plt.tick_params(axis='x', which='both', bottom=False)

plt.tight_layout()
plt.show()
```

```python
algorithms = data['Algorithm'].unique()

plt.figure(figsize=(12, 8))


for algorithm in algorithms:
    algorithm_data = data[data['Algorithm'] == algorithm]
    plt.plot(algorithm_data['Array Size'], algorithm_data['Comparisons'], marker='o', label=algorithm, linewidth=2)

plt.title('Nmero de comparaes por tamanho de array para diferentes algoritmos de ordenao')
plt.xlabel('Tamanho do Array')
plt.ylabel('Nmero de comparaes')
plt.legend()
plt.grid(True, axis='y')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tick_params(axis='x', which='both', bottom=False)

plt.tight_layout()
plt.show()
```
